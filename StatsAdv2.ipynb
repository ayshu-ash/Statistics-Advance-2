**Question 1: What is hypothesis testing in statistics?**

**Answer:**  

**1.Definition:**

Hypothesis Testing is a statistical method used to make decisions or inferences about a population parameter based on sample data.

* It involves formulating two competing statements (hypotheses) and using data to decide which one is more likely to be true.
* It provides a systematic way to test assumptions about population means, proportions, variances, etc.

 **2.Purpose:**

* To determine whether there is enough evidence in a sample to support or reject a claim about the population.
* To help researchers, scientists, and businesses make decisions under uncertainty.
* To avoid relying only on descriptive statistics by applying probability-based inference.

 **3.Steps in Hypothesis Testing:**

  1. **State the Hypotheses:**

       * **Null Hypothesis (H₀):** Assumes no effect or no difference.

       * **Alternative Hypothesis (H₁):** Assumes there is an effect or difference.

  2. **Choose Significance Level (α):**

       * Common choices: 0.05 (5%) or 0.01 (1%).

  3. **Select Test Statistic:**

       * Examples: *t-test, z-test, chi-square test, ANOVA*.

  4. **Make Decision Rule:**

       * Compare p-value with α or use critical values.

  5. **Draw Conclusion:**

       * If p-value < α → Reject H₀.
       * If p-value ≥ α → Fail to reject H₀.

  **4.Real-World Examples:**

  1. **Medical Research:** Testing if a new drug is more effective than the existing one.

        * H₀: The new drug has no effect compared to the old drug.
        * H₁: The new drug works better.

  2. **Business:** Testing if a new advertisement increases sales.

        * H₀: The advertisement does not affect sales.
        * H₁: The advertisement increases sales.

  3. **Education:** Checking if a new teaching method improves student performance.

        * H₀: No improvement in scores.
        * H₁: Scores improve with the new method.

 **5.Key Points**

  * Hypothesis testing always involves two hypotheses: H₀ and H₁.
  * Decisions are based on probability and evidence, not certainty.
  * Errors are possible:

     * **Type I Error:** Rejecting H₀ when it is true.
     * **Type II Error:** Failing to reject H₀ when it is false.
  * Provides the foundation for statistical inference in research and experiments.
  * Widely used in science, business, medicine, engineering, and social sciences.

**Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?**

**Answer:**

 **1.Definition of Null Hypothesis (H₀):**

  * The Null Hypothesis (H₀) is a statement that assumes no effect, no relationship, or no difference exists in the population.
  * It represents the default or original claim which researchers try to test.

   Example: "The average marks of male and female students are equal."

 **2.Definition of Alternative Hypothesis (H₁ or Ha):**

   * The Alternative Hypothesis (H₁) is a statement that contradicts the null hypothesis.
   * It assumes that there is an effect, a relationship, or a difference in the population.

   Example: "The average marks of male and female students are not equal."

 **3.Key Differences between H₀ and H₁:**

| **Aspect**              | **Null Hypothesis (H₀)**                        | **Alternative Hypothesis (H₁)**           |
| ----------------------- | ----------------------------------------------- | ----------------------------------------- |
| **Meaning**             | Assumes no effect, no difference, no change     | Assumes an effect, difference, or change  |
| **Role**                | Default assumption (status quo)                 | Competing claim to be proven              |
| **Decision Basis**      | Retained unless strong evidence against it      | Accepted if evidence supports it          |
| **Example (Drug test)** | "The new drug has no effect on patients."       | "The new drug improves patients' health." |
| **Example (Education)** | "New teaching method has no impact on results." | "New teaching method improves results."   |

 **4.Relation with Hypothesis Testing:**

   * Both H₀ and H₁ are formulated at the beginning of hypothesis testing.
   * Statistical tests are conducted to check whether sample data provides enough evidence to reject H₀ in favor of H₁.
   * Only one of them can be true at a time, but we never "accept" H₀ — we either reject it or fail to reject it.

 **5.Real-World Examples:**

   1. **Business (Advertisement)**

      * H₀: The new advertisement does not increase sales.
      * H₁: The new advertisement increases sales.

   2. **Medicine (Drug Trial)**

      * H₀: The new drug has no effect on blood pressure.
      * H₁: The new drug lowers blood pressure.

   3. **Education (Teaching Method)**

      * H₀: The new teaching method has no effect on student marks.
      * H₁: The new teaching method improves marks.

 **6.Key Points for Exams**

  *  Null Hypothesis (H₀) = no effect / no difference.
  * Alternative Hypothesis (H₁) = effect / difference exists.
  * Both are mutually exclusive and exhaustive.
  * Decision is based on p-value and significance level (α).
  * Central to all statistical inference and research testing.

**Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.**

**Answer:**

 **1.Definition of Significance Level (α):**

  * The significance level, denoted by **α**, is a pre-determined threshold used in hypothesis testing to decide whether to reject the null hypothesis (H₀)**.
  * It represents the probability of making a Type I error, i.e., rejecting H₀ when it is actually true.
  * Common values of α: 0.05 (5%), 0.01 (1%), 0.10 (10%).

 **2.Purpose / Significance:**

  * Controls the risk of wrong decisions in hypothesis testing.
  * Helps determine how strong the evidence must be before rejecting H₀.
  * Provides a clear criterion for decision making in statistical tests.
  * Balances the trade-off between Type I error (false positive) and Type II error (false negative).

 **3.Role in Deciding Test Outcome:**

   1. **Compute p-value**: Probability of observing the sample result (or more extreme) assuming H₀ is true.
   2. **Compare p-value with α**:

      * If **p-value ≤ α** → Reject H₀ (sufficient evidence against null).
      * If **p-value > α** → Fail to reject H₀ (insufficient evidence against null).

**Flow:**

Formulate H₀ and H₁ → Set α → Compute test statistic → Find p-value → Compare p-value with α → Make decision

 **4.Real-World Examples:**

  1. **Medical Research:**

       * α = 0.05, testing if a new drug lowers blood pressure.
       * If p-value = 0.03 → Reject H₀ → Drug has significant effect.

  2. **Business (Marketing Campaign):**

       * α = 0.01, testing if a new ad increases sales.
       * If p-value = 0.02 → Fail to reject H₀ → No strong evidence for ad effect.

  3. **Education:**

       * α = 0.05, testing if a new teaching method improves test scores.
       * If p-value = 0.07 → Fail to reject H₀ → Evidence is not strong enough.

 **5.Key Points:**

| **Aspect**        | **Details**                                          |
| ----------------- | ---------------------------------------------------- |
| **Symbol**        | α (alpha)                                            |
| **Represents**    | Probability of Type I error                          |
| **Common Values** | 0.01, 0.05, 0.10                                     |
| **Decision Rule** | Reject H₀ if p-value ≤ α; else fail to reject H₀     |
| **Role**          | Determines how strong evidence must be to reject H₀  |
| **Trade-off**     | Lower α → less Type I error but higher Type II error |

 **6.Why it is important:**

   * Ensures consistency and objectivity in hypothesis testing.
   * Helps researchers control error rates and make reliable decisions.
   * Directly influences conclusions drawn from experiments or data analysis.

**Question 4: What are Type I and Type II errors? Give examples of each.**

**Answer:**

 **1.Definition of Type I Error (False Positive):**

   * A Type I error occurs when the null hypothesis (H₀) is true, but we incorrectly reject it.
   * Probability of Type I error is denoted by α (significance level).
   * Also called a false positive because it indicates an effect that does not exist.

   **Example:**

  1. **Medical Test:**

      * H₀: A patient does not have a disease.
      * H₁: A patient has the disease.
      * Type I error → Test shows the patient has the disease when they actually do not.

 2. **Business:**

      * H₀: A new advertisement does not increase sales.
      * H₁: Advertisement increases sales.
      * Type I error → Concluding the ad works when it actually does not.

 **2.Definition of Type II Error (False Negative):**

* A Type II error occurs when the null hypothesis (H₀) is false, but we **fail to reject it.
* Probability of Type II error is denoted by **β**.
* Also called a false negative because it misses detecting a real effect.

  **Example:**

    1. **Medical Test:**

       * H₀: A patient does not have a disease.
       * H₁: A patient has the disease.
       * Type II error → Test shows the patient does not have the disease when they actually do.

    2. **Business:**

       * H₀: A new advertisement does not increase sales.
       * H₁: Advertisement increases sales.
       * Type II error → Concluding the ad does not work when it actually does.

 **3.Comparison of Type I and Type II Errors:**

| **Aspect**      | **Type I Error**                      | **Type II Error**                |
| --------------- | ------------------------------------- | -------------------------------- |
| **Also called** | False Positive                        | False Negative                   |
| **Occurs when** | H₀ is true, but rejected              | H₀ is false, but not rejected    |
| **Symbol**      | α (alpha)                             | β (beta)                         |
| **Consequence** | Detects an effect that does not exist | Misses an effect that exists     |
| **Example**     | Declaring a healthy patient sick      | Declaring a sick patient healthy |

 **4.Key Points**

  * Type I error → Rejecting a true null hypothesis.
  * Type II error → Failing to reject a false null hypothesis.
  * Trade-off:

      * Reducing α (Type I error) may increase β (Type II error) and vice versa.
  * Both errors are critical in research, medicine, and business decisions.
  * Understanding these errors helps in choosing significance levels and designing experiments.

**Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.**

**Answer:**

**1.Definition of Z-test:**

   * A Z-test is a statistical test used to determine whether there is a significant difference between sample and population means or between two sample means when the population standard deviation is known.
   * It is based on the standard normal distribution (Z-distribution).

  **Formula for single-sample Z-test:**

$$
Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
$$
       Where:

* $\bar{X}$ = Sample mean
* $\mu$ = Population mean
* $\sigma$ = Population standard deviation
* $n$ = Sample size

 **2.Definition of T-test:**

   * A T-test is a statistical test used to determine whether there is a significant difference between sample means or between a sample and population mean when the population standard deviation is unknown.
   * It is based on the Student's t-distribution, which adjusts for small sample sizes.

  **Formula for single-sample T-test:**

$$
t = \frac{\bar{X} - \mu}{s / \sqrt{n}}
$$

Where:

* $s$ = Sample standard deviation (used instead of population σ)

 **3.Key Differences between Z-test and T-test:**

| **Aspect**                        | **Z-test**                                                    | **T-test**                                                           |
| --------------------------------- | ------------------------------------------------------------- | -------------------------------------------------------------------- |
| **Distribution**                  | Standard normal (Z)                                           | Student’s t-distribution                                             |
| **Population Standard Deviation** | Known                                                         | Unknown                                                              |
| **Sample Size**                   | Large sample (n ≥ 30)                                         | Small sample (n < 30)                                                |
| **Use Case**                      | Testing means with known σ                                    | Testing means with unknown σ                                         |
| **Critical Value**                | Z-table                                                       | t-table (depends on degrees of freedom)                              |
| **Example**                       | Test if average height of 1000 adults equals 170 cm (σ known) | Test if average weight of 20 students differs from 60 kg (σ unknown) |

 **4.When to Use Each:**

  1. **Use Z-test when:**

        * Population standard deviation is known.
        * Sample size is large (n ≥ 30).
        * Example: Quality control in factories with known population variance.

  2. **Use T-test when:**

        * Population standard deviation is unknown.
        * Sample size is small (n < 30).
        * Example: Comparing average test scores of two small classes.

 **5.Types of T-tests**

  1. **One-sample t-test:** Compare sample mean with population mean.
  2. **Independent two-sample t-test:** Compare means of two independent groups.
  3. **Paired t-test:** Compare means of two related groups (before and after treatment).

 **6.Key Points**

  * Z-test → uses known population σ; T-test → uses sample s.
  * Z-test → for large samples; T-test → for small samples.
  * Both are used for hypothesis testing on means.
  * Correct choice ensures accurate statistical conclusions.
  * Z-test relies on normal distribution, T-test adjusts for uncertainty in small samples.

```python
#Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.

#(Include your Python code and output in the code box below.)

#Hint: Generate random number using random function.

#Answer:

import numpy as np
import matplotlib.pyplot as plt

def generate_binomial_data(n, p, size):
    """
    Generate random numbers following a binomial distribution.

    Args:
        n (int): Number of trials
        p (float): Probability of success in each trial
        size (int): Number of random samples to generate

    Returns:
        np.ndarray: Array of generated binomial data
    """
    return np.random.binomial(n=n, p=p, size=size)

def plot_histogram(data, bins=11, title="Binomial Distribution Histogram", xlabel="Number of Successes", ylabel="Frequency"):
    """
    Plot a histogram of the binomial data.

    Args:
        data (np.ndarray): Binomial data to plot
        bins (int): Number of bins
    """
    plt.figure(figsize=(10,6))
    plt.hist(data, bins=bins, color='lightgreen', edgecolor='black', rwidth=0.8)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xticks(range(0, max(data)+1))
    plt.grid(axis='y', alpha=0.75)
    plt.show()

# --- Main Program ---
if __name__ == "__main__":
    # Parameters
    n = 10        # Number of trials
    p = 0.5       # Probability of success
    size = 1000   # Number of random samples

    # Generate binomial data
    binomial_data = generate_binomial_data(n, p, size)

    # Print first 20 samples as example
    print("First 20 samples from binomial distribution:", binomial_data[:20])

    # Plot histogram
    plot_histogram(binomial_data)

```

```python
#Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.

"""sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
                        50.3, 50.4, 50.0, 49.7, 50.5, 49.9] """
#(Include your Python code and output in the code box below.)

#Answer:

import numpy as np
from scipy import stats

def perform_z_test(sample_data, population_mean, population_std, alpha=0.05):
    """
    Perform one-sample Z-test for the sample data.

    Args:
        sample_data (list or np.ndarray): Sample values
        population_mean (float): Population mean (μ)
        population_std (float): Population standard deviation (σ)
        alpha (float): Significance level (default 0.05)

    Returns:
        dict: Dictionary containing Z-statistic, p-value, and test conclusion
    """
    n = len(sample_data)
    sample_mean = np.mean(sample_data)

    # Compute Z-statistic
    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Two-tailed p-value
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

    # Decision
    if p_value < alpha:
        conclusion = "Reject H0: Significant difference exists."
    else:
        conclusion = "Fail to reject H0: No significant difference."

    return {
        "Z-statistic": z_stat,
        "p-value": p_value,
        "Conclusion": conclusion
    }

# --- Main Program ---
if __name__ == "__main__":
    # Sample dataset
    sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                   50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                   50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
                   50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

    # Population parameters (assumed known)
    population_mean = 50.0
    population_std = 0.8
    alpha = 0.05  # 5% significance level

    # Perform Z-test
    result = perform_z_test(sample_data, population_mean, population_std, alpha)

    # Display results
    print("Z-statistic:", round(result["Z-statistic"], 3))
    print("p-value:", round(result["p-value"], 4))
    print("Conclusion:", result["Conclusion"])
```

```python
#Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
#(Include your Python code and output in the code box below.)

#Answer:

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def simulate_normal_data(mean, std_dev, size):
    """
    Generate random numbers from a normal distribution.

    Args:
        mean (float): Mean of distribution
        std_dev (float): Standard deviation
        size (int): Number of data points

    Returns:
        np.ndarray: Generated data
    """
    return np.random.normal(loc=mean, scale=std_dev, size=size)

def compute_confidence_interval(data, confidence=0.95):
    """
    Compute confidence interval for the mean using t-distribution.

    Args:
        data (np.ndarray): Sample data
        confidence (float): Confidence level

    Returns:
        tuple: (mean, lower bound, upper bound)
    """
    n = len(data)
    mean = np.mean(data)
    sem = stats.sem(data)  # Standard error of the mean
    h = sem * stats.t.ppf((1 + confidence) / 2, n - 1)
    return mean, mean - h, mean + h

def plot_data(data, bins=30, title="Histogram of Normal Data", xlabel="Value", ylabel="Frequency"):
    """
    Plot histogram of data.

    Args:
        data (np.ndarray): Data to plot
    """
    plt.figure(figsize=(10,6))
    plt.hist(data, bins=bins, color='skyblue', edgecolor='black')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(axis='y', alpha=0.7)
    plt.show()

# --- Main Program ---
if __name__ == "__main__":
    # Parameters
    mean = 50
    std_dev = 5
    size = 1000

    # Simulate data
    data = simulate_normal_data(mean, std_dev, size)

    # Compute 95% confidence interval
    mean_val, lower, upper = compute_confidence_interval(data, confidence=0.95)
    print(f"Mean: {mean_val:.2f}")
    print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")

    # Plot histogram
    plot_data(data, bins=30, title="Normal Distribution of Simulated Data", xlabel="Value", ylabel="Frequency")
```

```python
#Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram.
#Explain what the Z-scores represent in terms of standard deviations from the mean.

#(Include your Python code and output in the code box below.)

#Answer:

import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate Z-scores for a dataset.

    Args:
        data (list or np.ndarray): Input data

    Returns:
        np.ndarray: Array of Z-scores
    """
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # Sample standard deviation
    z_scores = (data - mean) / std_dev
    return z_scores

def plot_histogram(data, bins=20, title="Histogram of Z-scores", xlabel="Z-score", ylabel="Frequency"):
    """
    Plot histogram of Z-scores.

    Args:
        data (np.ndarray): Z-score data
    """
    plt.figure(figsize=(10,6))
    plt.hist(data, bins=bins, color='lightcoral', edgecolor='black', rwidth=0.8)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(axis='y', alpha=0.7)
    plt.show()

# --- Main Program ---
if __name__ == "__main__":
    # Sample dataset
    data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
            50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
            50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
            50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

    # Calculate Z-scores
    z_scores = calculate_z_scores(np.array(data))

    # Display first 10 Z-scores as example
    print("First 10 Z-scores:", np.round(z_scores[:10], 3))

    # Plot histogram of Z-scores
    plot_histogram(z_scores)

    # Explanation
    print("\nExplanation:")
    print("Z-scores represent the number of standard deviations each data point is from the mean.")
    print("Positive Z-scores indicate values above the mean, negative Z-scores indicate values below the mean.")
```
